```python
import pandas as pd
```

```python
df = pd.read_csv('https://cocl.us/datascience_survey_data', index_col=0)
```

```python
df.sort_values('Very interested', ascending=False, inplace=True)
df_perc = round(df/2233, 2)
```

```python
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(20, 8))
plt.rc('font', size=14)
plt.rc('figure', titlesize=16)

color = ['#5cb85c', '#5bc0de', '#d9534f']

ax = df_perc.plot(kind='bar', figsize=(20, 8), title='Percentage of Respondents\' Interest in Data Science Areas', width=0.8, color=color)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.01, p.get_height() * 1.01))
```
