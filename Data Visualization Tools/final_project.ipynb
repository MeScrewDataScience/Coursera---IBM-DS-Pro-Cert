```python
import pandas as pd
```

```python
df = pd.read_csv('https://cocl.us/datascience_survey_data', index_col=0)
df.head()
```

```python
df.describe()
```

```python
df.sort_values('Very interested', ascending=False, inplace=True)
```

```python
df_perc = round(df/2233, 2)
df_perc
```

```python
import matplotlib.pyplot as plt
import numpy as np

xlabels = df_perc.index.tolist()
g1_perc = df_perc['Very interested']
g2_perc = df_perc['Somewhat interested']
g3_perc = df_perc['Not interested']

x = np.arange(len(xlabels))*3
width = 0.8

fig, ax = plt.subplots()
plt.figure(figsize=(20, 8))
rects1 = ax.bar(x - width, g1_perc, width, label='Very interested')
rects2 = ax.bar(x, g2_perc, width, label='Somewhat interested')
rects3 = ax.bar(x + width, g3_perc, width, label='Not interested')

# Add text to chart
ax.set_title('Percentage of Respondents\'s Interest in Data Science Areas')
ax.set_xticks(x)
ax.set_xticklabels(xlabels)


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 3, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
```
